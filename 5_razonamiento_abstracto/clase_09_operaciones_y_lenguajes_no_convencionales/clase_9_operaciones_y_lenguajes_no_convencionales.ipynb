Vamos paso a paso.  

**Definición de la operación**:  
$$
A * B = 
\begin{cases} 
A + B, & \text{si } A \geq 0 \\ 
A \times B, & \text{si } A < 0 
\end{cases}
$$

---



**1. Calcular $3 * (-5)$**  
Aquí $A = 3$ (≥ 0), así que usamos $A+B$:  
$$
3 * (-5) = 3 + (-5) = -2
$$

---



**2. Calcular \((-2) * (-3)\)**  
Aquí \(A = -2\) (< 0), así que usamos \(A \times B\):  
\[
(-2) * (-3) = (-2) \times (-3) = 6
\]

---



**3. Calcular \((-2) * (6)\)**  
Es decir, el resultado de \((3*(-5))\) es \(-2\), y el resultado de \((-2)*(-3)\) es \(6\), tenemos:  
\[
(-2) * 6
\]  
Aquí \(A = -2\) (< 0), usamos multiplicación:  
\[
(-2) * 6 = (-2) \times 6 = -12
\]

---



**Resultado final**: $-12$

La opción correcta es la **C**.
