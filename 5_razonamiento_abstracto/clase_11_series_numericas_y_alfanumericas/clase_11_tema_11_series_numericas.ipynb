````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

# Pregunta 1 

¡Claro que sí! Vamos a resolver este nuevo reto de la **Universidad de Antioquia**. Este es un problema de **progresiones aritméticas**, un tema fundamental en las pruebas de aptitud lógica.

---

### 1. Enunciado y Edición del Problema

**Tema 11: Series numéricas y alfa-numéricas**

**11.1. Considere la siguiente sucesión numérica:**
$$4, 9, 14, 19, 24, \dots$$

**Pregunta:**
> ¿Cuál es el elemento que ocupa la **posición 47** en la sucesión?

**Opciones de respuesta:**
(a) 234
(b) 189
(c) 201
(d) 249

---

### 2. Estrategia de Resolución "Flash" (Menos de 2 minutos)

Para resolver esto sin perder tiempo, identifica que la diferencia entre los números es constante:
* $9 - 4 = 5$
* $14 - 9 = 5$
* $19 - 14 = 5$

La diferencia común ($d$) es **5**. La fórmula mágica para cualquier término $n$ es:
$$a_n = a_1 + (n - 1) \cdot d$$

Sustituimos los valores para la posición **47**:
1. $a_1 = 4$ (primer término)
2. $d = 5$ (salto constante)
3. $n = 47$ (posición deseada)

$$a_{47} = 4 + (47 - 1) \cdot 5$$
$$a_{47} = 4 + (46 \cdot 5)$$
$$a_{47} = 4 + 230 = 234$$

**Respuesta Correcta: (a) 234.**

---



### 3. Implementación Didáctica con Python (Pandas, Seaborn e ipywidgets)

Este código te permitirá visualizar cómo crece la serie y verificar el resultado de forma interactiva.

````python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Crear el DataFrame con la progresión
posiciones = list(range(1, 48))
valores = [4 + (n - 1) * 5 for n in posiciones]
df = pd.DataFrame({'Posición': posiciones, 'Valor': valores})

# 2. Función de visualización
def visualizar_progresion(n_final):
    with out:
        clear_output(wait=True)
        # Filtrar datos hasta el paso seleccionado
        df_plot = df[df['Posición'] <= n_final]
        
        plt.figure(figsize=(10, 5))
        sns.set_style("whitegrid")
        
        # Gráfico de línea con puntos
        plot = sns.lineplot(data=df_plot, x='Posición', y='Valor', marker='o', color='teal')
        
        # Resaltar el último valor calculado
        ultimo_val = df_plot['Valor'].iloc[-1]
        plt.scatter(n_final, ultimo_val, color='red', s=100, zorder=5)
        plt.annotate(f'Pos {n_final}: {ultimo_val}', 
                     xy=(n_final, ultimo_val), 
                     xytext=(n_final-10, ultimo_val+20),
                     arrowprops=dict(facecolor='black', shrink=0.05))

        plt.title(f"Crecimiento de la Progresión Aritmética (d=5)", fontsize=15)
        plt.ylim(0, 260)
        plt.xlim(0, 50)
        plt.xticks(np.arange(0,50, 5))
        plt.yticks(np.arange(0,250, 40))
        plt.savefig(r"C:\Users\marco\Downloads\crecimiento_progresion_aritmetica.png")
        plt.show()

# 3. Widget interactivo
slider = widgets.IntSlider(value=1, min=1, max=47, step=1, 
                          description='Posición:', 
                          layout=widgets.Layout(width='50%'))
out = widgets.Output()

def on_value_change(change):
    visualizar_progresion(change['new'])

slider.observe(on_value_change, names='value')

display(slider, out)
# Carga inicial
visualizar_progresion(1)
````


### Resumen de aprendizaje
* **Diferencia constante ($d$):** Es lo que sumas en cada paso.
* **Linealidad:** Al ser una suma constante, si graficas los resultados, siempre obtendrás una línea recta. Por eso usamos `sns.lineplot`.
* **Cálculo mental rápido:** Si $d=5$, el término $n$ siempre terminará en el mismo dígito que $a_1$ o en su variante (en este caso, todos terminan en 4 o 9). Como 47 es impar, y empezamos en 4, el resultado debía terminar en 4.

---

### 1. Enunciado y Edición del Problema

**Tema 11: Series y sucesiones**

**11.2. En la sucesión:**
$$4, \frac{10}{3}, \frac{8}{3}, 2, \square, \square, \dots$$

**Pregunta:**
> ¿Cuál es el **producto** de los 2 términos que siguen en los espacios marcados con $\square$?

**Opciones de respuesta:**
(a) $8/3$
(b) $8/9$
(c) $0$
(d) $-4/9$

---

### 2. Estrategia de Resolución "Flash" (Menos de 2 minutos)

La clave en las sucesiones con fracciones es **homogeneizar los denominadores**. Vamos a expresar todos los términos con denominador $3$:

1.  Primer término ($a_1$): $4 = \frac{12}{3}$
2.  Segundo término ($a_2$): $\frac{10}{3}$
3.  Tercer término ($a_3$): $\frac{8}{3}$
4.  Cuarto término ($a_4$): $2 = \frac{6}{3}$

**Paso 1: Identificar el patrón**
Observamos los numeradores: $12, 10, 8, 6 \dots$ Se restan $2$ en cada paso. La diferencia común es $d = -\frac{2}{3}$.

**Paso 2: Calcular los términos faltantes**
* Siguiente término ($a_5$): $\frac{6}{3} - \frac{2}{3} = \frac{4}{3}$
* Siguiente término ($a_6$): $\frac{4}{3} - \frac{2}{3} = \frac{2}{3}$

**Paso 3: Calcular el producto**
$$\text{Producto} = \frac{4}{3} \cdot \frac{2}{3} = \frac{8}{9}$$

**Respuesta Correcta: (b) 8/9.**

---

### 3. Implementación Didáctica con Python

Este script utiliza **Pandas** para manejar la serie y **Matplotlib/Seaborn** para mostrar la progresión decreciente.

```python


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Creación de los datos
n = list(range(1, 7))
# a_n = 4 + (n-1)*(-2/3) -> simplificado: (14 - 2n) / 3
valores = [(14 - 2*i)/3 for i in n]
etiquetas = ["12/3", "10/3", "8/3", "6/3", "4/3", "2/3"]

df = pd.DataFrame({'Posición': n, 'Valor': valores, 'Etiqueta': etiquetas})

# 2. Función de animación/visualización
def animar_sucesion(pasos):
    with out:
        clear_output(wait=True)
        df_plot = df.iloc[:pasos]
        
        plt.figure(figsize=(9, 5))
        sns.set_style("darkgrid")
        
        # Graficamos la tendencia
        ax = sns.regplot(data=df_plot, x='Posición', y='Valor', 
                         scatter_kws={'s':150, 'color':'orange'}, 
                         line_kws={'color':'blue', 'alpha':0.3})
        
        # Etiquetas de las fracciones
        for i, row in df_plot.iterrows():
            plt.text(row['Posición'], row['Valor'] + 0.1, row['Etiqueta'], 
                     ha='center', fontsize=12, fontweight='bold')
            
        plt.title("Visualización de la Sucesión Aritmética Decreciente", fontsize=14)
        plt.ylim(0, 4.5)
        plt.xlim(0.5, 6.5)
        plt.xticks(n)
        plt.show()

# 3. Widget interactivo
slider = widgets.IntSlider(value=4, min=1, max=6, step=1, description='Términos:')
out = widgets.Output()

def on_change(change):
    animar_sucesion(change['new'])

slider.observe(on_change, names='value')
display(slider, out)
animar_sucesion(4)
````

### Análisis de la Razón
En este problema, la razón es una **constante negativa** ($d = -2/3$). En el gráfico de Seaborn verás una pendiente negativa perfecta. 

* **Tip Didáctico:** Siempre que veas fracciones y enteros mezclados en una serie, convierte los enteros a fracciones equivalentes. Esto hace que el patrón de los numeradores "salte" a la vista de inmediato.


# Pregunta 3  

Para cerrar este ciclo de retos, resolveremos la **pregunta 11.3**, que es un fascinante ejercicio de **promedios ponderados y secuencias piramidales**.

---

### 1. Enunciado y Edición del Problema

**11.3. La figura representa una pared formada por 15 ladrillos.**

En los ladrillos de la base se escriben, en orden, los primeros **cinco números pares positivos**. En cada ladrillo de las hiladas superiores se escribe el número correspondiente al **promedio** de los números escritos en los dos ladrillos que lo sustentan (los que están justo debajo).

**Pregunta:**
> ¿Cuál es la suma de **todos** los números escritos en los 15 ladrillos?

**Opciones de respuesta:**
(a) 85
(b) 90
(c) 100
(d) 50

---

### 2. Estrategia de Resolución "Flash" (Menos de 2 minutos)

**Paso 1: Identificar la base.**
Los primeros cinco números pares positivos son:
$$[2, 4, 6, 8, 10]$$
La suma de la base es $2+4+6+8+10 = 30$. El promedio de la base es $6$.

**Paso 2: Observar la simetría.**
Cuando calculas promedios de una progresión aritmética simétrica (como los pares), el valor central se mantiene y la pirámide tiende a converger hacia el promedio global de la base.
* **Fila 2:** $\frac{2+4}{2}=3, \frac{4+6}{2}=5, \frac{6+8}{2}=7, \frac{8+10}{2}=9 \rightarrow [3, 5, 7, 9]$ (Suma = 24)
* **Fila 3:** $\frac{3+5}{2}=4, \frac{5+7}{2}=6, \frac{7+9}{2}=8 \rightarrow [4, 6, 8]$ (Suma = 18)
* **Fila 4:** $\frac{4+6}{2}=5, \frac{6+8}{2}=7 \rightarrow [5, 7]$ (Suma = 12)
* **Fila 5 (Cúspide):** $\frac{5+7}{2}=6 \rightarrow [6]$ (Suma = 6)

**Paso 3: Suma total.**
Suma de las filas: $30 + 24 + 18 + 12 + 6 = \mathbf{90}$.

**Respuesta Correcta: (b) 90.**

---

### 3. Implementación Didáctica con Python

Usaremos **Pandas** para construir la pirámide y **Seaborn** para visualizar la "calidez" de los números (Heatmap), viendo cómo convergen al centro.



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Lógica de Cálculo de la Pirámide
def calcular_piramide():
    filas = [[2, 4, 6, 8, 10]]  # Base
    while len(filas[-1]) > 1:
        ant = filas[-1]
        nueva = [(ant[i] + ant[i+1])/2 for i in range(len(ant)-1)]
        filas.append(nueva)
    return filas

datos_piramide = calcular_piramide()

# 2. Función de Dibujo Geométrico (Isósceles)
def dibujar_pared(niveles_visibles):
    with out:
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(10, 6))
        
        suma_total = 0
        ancho_ladrillo = 1.0
        alto_ladrillo = 0.5
        
        # Dibujamos de abajo hacia arriba
        for i in range(niveles_visibles):
            fila_valores = datos_piramide[i]
            n_ladrillos = len(fila_valores)
            suma_total += sum(fila_valores)
            
            # El offset X centra la fila: (TotalBase - LadrillosActuales) / 2
            offset_x = (len(datos_piramide[0]) - n_ladrillos) / 2.0
            
            for j, val in enumerate(fila_valores):
                x = offset_x + j
                y = i * alto_ladrillo
                
                # Crear el ladrillo (Rectángulo)
                rect = patches.Rectangle((x, y), ancho_ladrillo, alto_ladrillo, 
                                         linewidth=2, edgecolor='brown', 
                                         facecolor='#e28743', alpha=0.8)
                ax.add_patch(rect)
                
                # Añadir el número dentro del ladrillo
                ax.text(x + ancho_ladrillo/2, y + alto_ladrillo/2, f'{val:g}', 
                        ha='center', va='center', fontweight='bold', fontsize=12)

        # Configuración de los ejes
        ax.set_xlim(0, 5)
        ax.set_ylim(0, 2.7)
        ax.set_aspect('equal')
        ax.axis('off')
        
        plt.title(f"Pared de Ladrillos (Promedios)\nSuma Acumulada: {int(suma_total)}", 
                  fontsize=16, pad=20, color='#873e23')
        plt.show()

# 3. Interfaz con ipywidgets
slider = widgets.IntSlider(value=1, min=1, max=5, step=1, 
                          description='Construir:', 
                          style={'description_width': 'initial'})
out = widgets.Output()

def update(change):
    dibujar_pared(change['new'])

slider.observe(update, names='value')
display(slider, out)

# Estado inicial
dibujar_pared(1)
````

```

# Pregunta 4  

Para este último reto, nos enfocaremos en la **pregunta 11.4** de la imagen `pregunta_4.jpeg`. Este es un problema de **progresiones aritméticas** aplicado a una situación física de corte de materiales.

---

### 1. Enunciado y Edición del Problema

**11.4. Problema del Escultor y la Barra de Metal**

Un escultor desea cortar **10 trozos** de una barra de metal para ubicarlos verticalmente uno tras otro, de tal manera que cada trozo después del primero mide **2 metros más** que el anterior. Si la longitud total de la barra a cortar es de **100 metros** y se usa toda la barra, entonces:

**Pregunta:**
> ¿Cuál es la longitud del **octavo trozo**, expresada en metros?

**Opciones de respuesta:**
(a) 5
(b) 15
(c) 19
(d) 90 
*(Nota: Al resolver, verificaremos si las opciones coinciden con el cálculo matemático exacto).*

---

### 2. Estrategia de Resolución "Flash" (Menos de 2 minutos)

Este problema se resuelve usando la fórmula de la **suma de una progresión aritmética**:
$$S_n = \frac{n}{2} [2a_1 + (n - 1)d]$$

Donde:
* $S_n = 100$ (Longitud total de la barra).
* $n = 10$ (Número de trozos).
* $d = 2$ (Diferencia de longitud entre trozos).
* $a_1 = ?$ (Longitud del primer trozo).

**Paso 1: Hallar $a_1$**
$$100 = \frac{10}{2} [2a_1 + (10 - 1)2]$$
$$100 = 5 [2a_1 + 18]$$
$$20 = 2a_1 + 18$$
$$2 = 2a_1 \implies a_1 = 1$$

**Paso 2: Hallar el octavo trozo ($a_8$)**
$$a_n = a_1 + (n - 1)d$$
$$a_8 = 1 + (8 - 1)2$$
$$a_8 = 1 + (7 \times 2) = 1 + 14 = 15$$

**Respuesta Correcta: (b) 15.**

---



### 3. Implementación Didáctica con Python

Usaremos **Pandas** para tabular cada trozo y **Matplotlib/Seaborn** para visualizar la "escalera" que forman los trozos de metal.


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Generar los datos de los 10 trozos
def calcular_trozos(a1=1):
    n = list(range(1, 11))
    longitudes = [a1 + (i-1)*2 for i in n]
    return pd.DataFrame({'Trozo': n, 'Longitud (m)': longitudes})

df_trozos = calcular_trozos()
df_trozos
````

````python

# 2. Visualización Animada
def visualizar_escultura(hasta_n):
    with out:
        clear_output(wait=True)
        df_plot = df_trozos.iloc[:hasta_n]
        
        plt.figure(figsize=(10, 5))
        sns.set_style("white")
        
        # Dibujar los trozos como barras verticales (escultura)
        colors = ['teal' if i != 8 else 'orange' for i in range(1, hasta_n + 1)]
        barplot = sns.barplot(data=df_plot, x='Trozo', y='Longitud (m)', legend=False)
        
        # Añadir etiquetas sobre las barras
        for i, val in enumerate(df_plot['Longitud (m)']):
            plt.text(i, val + 0.5, f"{val}m", ha='center', fontweight='bold')
            
        plt.axhline(df_trozos['Longitud (m)'].mean(), color='red', linestyle='--', label='Promedio')
        plt.title(f"Visualización de la Escultura: Corte de {hasta_n} trozos\nSuma acumulada: {df_plot['Longitud (m)'].sum()}m", fontsize=14)
        plt.ylim(0, 22)
        plt.legend()
        plt.show()

# 3. Widget interactivo
slider = widgets.IntSlider(value=1, min=1, max=10, description='Ver trozos:')
out = widgets.Output()

def on_change(change):
    visualizar_escultura(change['new'])

slider.observe(on_change, names='value')
display(slider, out)
visualizar_escultura(1)
````

### Tips para el examen:
1.  **El término medio:** En una progresión aritmética de 10 términos que suma 100, el promedio es $100/10 = 10$. El promedio de una serie aritmética siempre está entre los términos centrales ($a_5$ y $a_6$).
2.  **Verificación rápida:** Si el promedio es 10, entonces $a_5 + a_6$ debe ser 20. Como $d=2$, los términos centrales son 9 y 11. De ahí puedes contar hacia adelante: $a_7=13, a_8=15$. ¡Llegaste a la respuesta sin fórmulas complejas!

### Implementing MLPs with Keras

Keras is a high-level Deep Learning API that allows you to easily build, train, evaluate, and execute all sorts of neural networks. Its documentation (or specification) is available at https://keras.io/. The reference implementation, also called Keras, was developed by François Chollet as part of a research project and was released as an open source project in
March 2015. It quickly gained popularity, owing to its ease of use, flexibility, and beautiful design. To perform the heavy computations required by neural networks, this reference implementation relies on a computation backend. At present, you can choose from three popular open source Deep Learning libraries: TensorFlow, Microsoft Cognitive Toolkit(CNTK), and Theano. Therefore, to avoid any confusion, we will refer to this reference implementation as multibackend Keras. Since late 2016, other implementations have been released. You can now run Keras on Apache MXNet, Apple’s Core ML, JavaScript or TypeScript (to run Keras code in a web browser), and PlaidML (which can run on all sorts of GPU devices, not just Nvidia). Moreover, TensorFlow itself now comes bundled with its own Keras implementation, tf.keras. It only supports TensorFlow as the backend, but it has the advantage of offering some very useful extra features (see Figure 10-10): for example, it supports TensorFlow’s Data API, which makes it easy to load and preprocess data efficiently. For this reason, we will use tf.keras in this book. However, in this chapter we will not use any of the TensorFlow-specific features, so the code should run fine on other Keras implementations as well (at least in Python), with only minor modifications, such as changing the imports.
