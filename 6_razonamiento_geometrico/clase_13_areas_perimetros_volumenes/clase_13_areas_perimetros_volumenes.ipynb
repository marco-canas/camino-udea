# Idea clave (atajo mental)

En un paralelogramo:

* las bases son paralelas
* la altura es la misma
* el área depende de la base

- Entonces, **las áreas son proporcionales a las bases** cuando comparten la misma altura.

---



# Análisis rápido del problema

Tenemos:

* $PQRS$ es un paralelogramo
* $T$ es el punto medio de $SR$
* queremos el área de $\triangle QST$ respecto al paralelogramo

---



# Estrategia rápida

1. El área del paralelogramo es:

$$
\text{Área} = SR \times h
$$

---



2. El triángulo $\triangle QST$ tiene:

* base = $ST$
* altura = la misma del paralelogramo

Pero:

$$
ST = \frac{SR}{2}
$$

---



3. Área del triángulo:

$$
\frac{1}{2} \times ST \times h
$$

---



Sustituyendo:

$$
\frac{1}{2} \times \frac{SR}{2} \times h
$$

---

$$
= \frac{SR \cdot h}{4}
$$

---



4. Comparando con el paralelogramo:

$$
\frac{\text{Área triángulo}}{\text{Área paralelogramo}} = \frac{1}{4}
$$

---



#  Respuesta correcta:

**C) $1/4$**

---



#  Estrategia tipo examen (ultra rápida)

* punto medio → divide base en 2
* triángulo → mitad del rectángulo/paralelogramo
* mitad de la mitad →

$$
1/4
$$

---



# Representación con matplotlib

```python


````python
import matplotlib.pyplot as plt

# Puntos
P = (0,2)
Q = (4,2)
R = (5,0)
S = (1,0)
T = ((S[0]+R[0])/2, (S[1]+R[1])/2)

# Paralelogramo
x = [P[0], Q[0], R[0], S[0], P[0]]
y = [P[1], Q[1], R[1], S[1], P[1]]

plt.plot(x,y,'k-')

# Triángulo QST
plt.plot([Q[0],S[0],T[0],Q[0]],
         [Q[1],S[1],T[1],Q[1]],
         'r-')

# Puntos
for punto,nombre in zip([P,Q,R,S,T],["P","Q","R","S","T"]):
    plt.scatter(*punto)
    plt.text(punto[0]+0.1,punto[1]+0.1,nombre)

plt.title("Paralelogramo y triángulo QST")
plt.axis('equal')
plt.grid()
plt.savefig(r"C:\Users\marco\Downloads\paralelogramo_triangulo_matplotlib.png")
plt.show()
````

#  PROBLEMA 1 (tipo paralelogramo + punto medio)

En un paralelogramo $ABCD$, $M$ es el punto medio de $BC$. Se traza el triángulo $ADM$.

¿Qué fracción del área del paralelogramo representa $\triangle ADM$?

---


````python
import matplotlib.pyplot as plt

A=(0,2)
B=(4,2)
C=(5,0)
D=(1,0)
M=((B[0]+C[0])/2,(B[1]+C[1])/2)

# Dibujar paralelogramo
plt.plot([A[0],B[0],C[0],D[0],A[0]],
         [A[1],B[1],C[1],D[1],A[1]])

# Dibujar triángulo
plt.plot([A[0],D[0],M[0],A[0]],
         [A[1],D[1],M[1],A[1]])

# Dibujar puntos
plt.scatter(*zip(A,B,C,D,M))

# Agregar etiquetas (nombres)
puntos = {'A':A, 'B':B, 'C':C, 'D':D, 'M':M}

for nombre, (x,y) in puntos.items():
    plt.text(x+0.1, y+0.1, nombre, fontsize=12)

# Configuración
plt.axis('equal')
plt.grid()

# Guardar imagen
plt.savefig(r"C:\Users\marco\Downloads\prpoblema_triangulo_paralelogramo_matplotlib.png")

plt.show()
````


## Estrategia rápida

* misma altura
* base del triángulo: $AD$
* Área del triángulo = mitad $(AD \cdot h)$ = mitad (área del paralelo gramo)

---



## Solución

Área del paralelogramo:

$$
A_{\text{paralelogramo}} = AD \cdot h
$$

Área del triángulo:

$$
A_{\triangle} = \frac{1}{2} \cdot AD \cdot h = \frac{1}{2} A_{\text{paralelogramo}}
$$

---



## Respuesta

$$
\boxed{1/2}
$$







## plotly

Keras is a high-level Deep Learning API that allows you to easily build, train, evaluate, and execute all sorts of neural networks. Its documentation (or specification) is available at https://keras.io/. The reference implementation, also called Keras, was developed by François Chollet as part of a research project and was released as an open source project in
March 2015. It quickly gained popularity, owing to its ease of use, flexibility, and beautiful design. To perform the heavy computations required by neural networks, this reference implementation relies on a computation backend. At present, you can choose from three popular open source Deep Learning libraries: TensorFlow, Microsoft Cognitive Toolkit(CNTK), and Theano. Therefore, to avoid any confusion, we will refer to this reference implementation as multibackend Keras. Since late 2016, other implementations have been released. You can now run Keras on Apache MXNet, Apple’s Core ML, JavaScript or TypeScript (to run Keras code in a web browser), and PlaidML (which can run on all sorts of GPU devices, not just Nvidia). Moreover, TensorFlow itself now comes bundled with its own Keras implementation, tf.keras. It only supports TensorFlow as the backend, but it has the advantage of offering some very useful extra features (see Figure 10-10): for example, it supports TensorFlow’s Data API, which makes it easy to load and preprocess data efficiently.
