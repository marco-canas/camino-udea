````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img

# Ejemplo de uso:
# generar_qr("https://www.ejemplo.com")

````

### Enunciado del Problema (Editado y Completado)

**Tema 13: Relaciones espaciales**

**13.1. La figura mostrada se ha construido uniendo 7 cubos unitarios.**

Se trata de un "núcleo" central con cubos adyacentes en cada una de sus caras (una cruz tridimensional o estructura de ejes $x, y, z$).

**Pregunta:**
> La razón del **volumen** (en unidades cúbicas) a la **superficie** (en unidades cuadradas) de esta figura es:

**Opciones:**
(a) $1/6$
(b) $7/30$
(c) $7/36$
(d) $1/5$

---



### Estrategia de Resolución "Flash" (Menos de 2 minutos)

Para hallar la razón $\frac{\text{Volumen}}{\text{Área Superficial}}$, seguimos estos pasos:

1.  **Cálculo del Volumen ($V$):**
    * El enunciado nos dice que hay **7 cubos unitarios**.
    * Como cada cubo tiene volumen $1 \times 1 \times 1 = 1$, el volumen total es simplemente la cantidad de cubos.
    * **$V = 7$**

2.  **Cálculo del Área Superficial ($A$):**
    * Cada cubo tiene **6 caras**. Si estuvieran separados, habría $7 \times 6 = 42$ caras.
    * Sin embargo, al unirlos, algunas caras quedan ocultas (se "tocan").
    * **Análisis del núcleo:** El cubo central está tocando a otros cubos por sus **6 caras** (arriba, abajo, frente, atrás, izquierda, derecha). Por lo tanto, el cubo central aporta **0 caras** a la superficie exterior.
    * **Análisis de los brazos:** Hay 6 cubos externos. Cada uno de ellos está pegado al núcleo por **1 cara**. Entonces, cada uno muestra $6 - 1 = 5$ caras al exterior.
    * Área total: $6 \text{ cubos} \times 5 \text{ caras} = \mathbf{30}$.
    * **$A = 30$**

3.  **Razón Final:**
    * $\text{Razón} = \frac{V}{A} = \frac{7}{30}$

**Respuesta Correcta: (b) 7/30.**

---

### Implementación Didáctica en Python

Usaremos un enfoque de "vóxeles" para representar la figura en 3D y calcular sus propiedades automáticamente.




````python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# 1. Definición de la estructura (Coordenadas de los 7 cubos)
# El centro es (0,0,0)
cubos = np.array([
    [0,0,0], # Centro
    [1,0,0], [-1,0,0], # Eje X
    [0,1,0], [0,-1,0], # Eje Y
    [0,0,1], [0,0,-1]  # Eje Z
])

def calcular_razon():
    volumen = len(cubos)
    # Cada cubo tiene 6 caras. Restamos 2 por cada contacto entre cubos.
    # En esta estructura hay 6 contactos (uno por cada brazo unido al centro).
    contactos = 6 
    caras_totales = volumen * 6
    area = caras_totales - (contactos * 2) 
    return volumen, area, volumen/area

# 2. Visualización Interactiva
def plot_3d_cubes(elev=20, azim=45, mostrar_nucleo=True):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Datos para representar los cubos (Voxels)
    voxels = np.zeros((3, 3, 3), dtype=bool)
    # Trasladamos las coordenadas de [-1, 1] a [0, 2] para la matriz
    for c in cubos:
        voxels[c[0]+1, c[1]+1, c[2]+1] = True
        
    # Colores: Núcleo distinto a los brazos
    colors = np.empty(voxels.shape, dtype=object)
    colors[1, 1, 1] = '#FF5555' # Núcleo Rojo
    for c in cubos[1:]:
        colors[c[0]+1, c[1]+1, c[2]+1] = '#5555FF88' # Brazos Azules transparentes

    ax.voxels(voxels, facecolors=colors, edgecolor='k', linewidth=0.5)
    
    ax.set_title(f"Visualización de la Estructura (7 Cubos)\nVolumen = 7 | Área = 30", fontsize=14)
    ax.view_init(elev=elev, azim=azim)
    ax.axis('off')
    plt.show()

# Widget interactivo para rotar la figura
slider_elev = widgets.IntSlider(value=20, min=-90, max=90, description='Elevación:')
slider_azim = widgets.IntSlider(value=45, min=0, max=360, description='Rotación:')

ui = widgets.HBox([slider_elev, slider_azim])
out = widgets.interactive_output(plot_3d_cubes, {'elev': slider_elev, 'azim': slider_azim})

display(ui, out)

# 3. Resumen en Pandas para el reporte
v, a, r = calcular_razon()
df_resumen = pd.DataFrame({
    "Métrica": ["Volumen (V)", "Área Superficial (A)", "Razón (V/A)"],
    "Valor": [float(v), float(a), f"{v}/{a} ({round(r, 4)})"]
})

print("\nAnálisis de la Razón Espacial:")
display(df_resumen)
````

### ¿Por qué esta estrategia es ganadora?

* **El método de la "Resta de Contactos":** En lugar de contar caras una por una, recuerda: **Cada vez que dos cubos se pegan, desaparecen 2 caras del área superficial** (una de cada cubo). Como hay 6 uniones, restamos $6 \times 2 = 12$ caras al total teórico de 42. $42 - 12 = 30$.
* **Visualización mental:** Imagina que pintas la figura. Solo pintas las caras de afuera. El cubo del centro está totalmente "seco", no tiene pintura. Los otros 6 tienen una cara pegada (seca) y 5 caras pintadas. $6 \times 5 = 30$.

### Implementing MLPs with Keras

Keras is a high-level Deep Learning API that allows you to easily build, train, evaluate, and execute all sorts of neural networks. Its documentation (or specification) is available at https://keras.io/. The reference implementation, also called Keras, was developed by François Chollet as part of a research project and was released as an open source project in
March 2015. It quickly gained popularity, owing to its ease of use, flexibility, and beautiful design. To perform the heavy computations required by neural networks, this reference implementation relies on a computation backend. At present, you can choose from three popular open source Deep Learning libraries: TensorFlow, Microsoft Cognitive Toolkit(CNTK), and Theano. Therefore, to avoid any confusion, we will refer to this reference implementation as multibackend Keras. Since late 2016, other implementations have been released. You can now run Keras on Apache MXNet, Apple’s Core ML, JavaScript or TypeScript (to run Keras code in a web browser), and PlaidML (which can run on all sorts of GPU devices, not just Nvidia). Moreover, TensorFlow itself now comes bundled with its own Keras implementation, tf.keras. It only supports TensorFlow as the backend, but it has the advantage of offering some very useful extra features (see Figure 10-10): for example, it supports TensorFlow’s Data API, which makes it easy to load and preprocess data efficiently. For this reason, we will use tf.keras in this book. However, in this chapter we will not use any of the TensorFlow-specific features, so the code should run fine on other Keras implementations as well (at least in Python), with only minor modifications, such as changing the imports.
