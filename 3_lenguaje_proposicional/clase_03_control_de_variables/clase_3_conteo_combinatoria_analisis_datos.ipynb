# Problema 1 (bolitas de chicle)

3 rojas
2 azules

Se sacan **3 bolas**.

¿Qué se puede afirmar con certeza?

---

### Estrategia

Enumerar combinaciones.

RRR
RRB
RBB

Siempre aparece una roja.

---

✔ **Respuesta: al menos una roja**

---

### Verificación con Python

```python
import itertools

bolas = ["R","R","R","A","A"]

comb = set()

for c in itertools.combinations(bolas,3):
    comb.add(tuple(sorted(c)))

print(comb)
```

Contar pares posibles.

---

### Solución con Python

```python
contador = 0

for d in range(1,10):
    for u in range(10):
        if d > u:
            contador += 1

contador
```

# Problema 5 (galletas)

Si cada persona come 5 → sobran 4

Si cada persona come 6 → faltan 7

---

### Estrategia

Sistema de ecuaciones.

---

### Python

```python
import sympy as sp

p,g = sp.symbols('p g')

eq1 = g - (5*p+4)
eq2 = g - (6*p-7)

sp.solve((eq1,eq2),(p,g))
```

---

✔ **59 galletas**

# Problema 6 (triángulo numérico)

Cada número es diferencia de los dos inferiores.

---

### Estrategia

Calcular desde abajo hacia arriba.

---

### Python

```python
import numpy as np

fila=[7,9,4]

nivel2=[abs(fila[i]-fila[i+1]) for i in range(len(fila)-1)]

nivel3=abs(nivel2[0]-nivel2[1])

nivel3
```
