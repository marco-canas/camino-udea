```python
import numpy as np 
import pandas as pd 
```

```python
respuestas = pd.read_csv('respuestas_grupo_1.csv') 
respuestas.head()   
```

```python
for i in range(1,13):
    respuestas['correctas_est_'+str(i)] = respuestas['answer_right'] == respuestas['resp_'+str(i)]
```

```python
respuestas.head() 
```

```python
calificacion_diagnostico = []
for i in range(1,13):
    calificacion_diagnostico.append(respuestas['correctas_est_'+str(i)].sum()/38*5)

calificacion_diagnostico 
```
