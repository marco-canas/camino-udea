````python
import qrcode
from IPython.display import display
from PIL import Image

def generar_qr(url, version=1, box_size=4, border=4, fill_color="black", back_color="white"):
    """
    Genera y muestra un código QR para una URL dada en un Jupyter Notebook.
    
    Parámetros:
    - url: La URL para la cual generar el código QR
    - version: Tamaño del QR (1-40, donde 1 es el más pequeño)
    - box_size: Número de píxeles por cada "caja" del QR
    - border: Grosor del borde blanco alrededor del QR (en cajas)
    - fill_color: Color del código QR
    - back_color: Color de fondo del código QR
    
    Retorna:
    - Muestra el código QR directamente en el notebook
    - Retorna el objeto QR generado
    """
    # Configurar el generador de QR
    qr = qrcode.QRCode(
        version=version,
        error_correction=qrcode.constants.ERROR_CORRECT_L,
        box_size=box_size,
        border=border,
    )
    
    # Añadir la URL al QR
    qr.add_data(url)
    qr.make(fit=True)
    
    # Crear la imagen del QR
    img = qr.make_image(fill_color=fill_color, back_color=back_color)
     
    return img
````

---



### 1. Enunciado del Problema (5.1)

> "Un grupo de **8 obreros**, los cuales trabajan todos con la misma eficiencia, ejecuta una cierta obra trabajando durante **20 días**. Si solo se contara con **2 obreros** de ese grupo, el número de días que necesitarían estos 2 obreros en ejecutar la misma obra es:"

---



### 2. Estrategia Relámpago (< 2 min)

Este es un problema de **proporcionalidad inversa**: a *menor* cantidad de obreros, se requiere *mayor* cantidad de tiempo para terminar la misma obra.

**Estrategia de la Constante de Trabajo ($W$):**
El trabajo total se puede medir en "obreros-día".
1.  **Calcular el trabajo total:** $8 \text{ obreros} \times 20 \text{ días} = 160 \text{ unidades de trabajo}$.
2.  **Dividir por la nueva cantidad:** $\frac{160 \text{ unidades}}{2 \text{ obreros}} = 80 \text{ días}$.



**Estrategia de la Razón:**
* La cantidad de obreros se redujo a la **cuarta parte** ($8 \div 4 = 2$).
* Por lo tanto, el tiempo debe **cuadruplicarse** ($20 \times 4 = 80$).

**Respuesta Correcta: (d) 80**.

---



### 3. Modelado y Visualización con Python

Utilizaremos **Pandas** para tabular la relación y **Matplotlib** para graficar la curva de proporcionalidad inversa ($y = \frac{k}{x}$).


````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Definición de la constante de trabajo k (Obreros * Días)
k = 8 * 20 

# Crear datos para la curva
obreros = np.array([1, 2, 4, 8, 10, 16])
dias = k / obreros

df = pd.DataFrame({'Obreros': obreros, 'Días': dias})

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(df['Obreros'], df['Días'], marker='o', linestyle='--', color='teal')
plt.title('Relación Inversamente Proporcional: Obreros vs Tiempo')
plt.xlabel('Número de Obreros')
plt.ylabel('Días Requeridos')

# Resaltar el punto del problema
plt.annotate(f'Solución: (2, 80)', xy=(2, 80), xytext=(4, 100),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.grid(True, linestyle=':', alpha=0.6)
plt.savefig(r"C:\Users\marco\Downloads\proporcionalidad_inversa.png")  # Guardar la figura como imagen
plt.show()

print(df)
````

### 4. Interactividad para Aprendizaje (ipywidgets)

Con este control interactivo, puedes observar cómo cambia el tiempo requerido al variar la cantidad de obreros, manteniendo la eficiencia constante.

````python
import ipywidgets as widgets

def calcular_tiempo(n_obreros):
    trabajo_total = 160
    tiempo = trabajo_total / n_obreros
    print(f"--- Simulación de Obra ---")
    print(f"Si usamos {n_obreros} obreros, terminarán en {tiempo:.1f} días.")
    print(f"Constante de Trabajo: {n_obreros} * {tiempo:.1f} = 160")

widgets.interact(calcular_tiempo, n_obreros=widgets.IntSlider(min=1, max=40, step=1, value=8))
````

### Resumen Matemático
La relación inversa se expresa como:
$$x_1 \cdot y_1 = x_2 \cdot y_2$$
Sustituyendo los valores del problema:
$$8 \cdot 20 = 2 \cdot x$$
$$160 = 2x \implies x = \frac{160}{2} = 80$$

Vamos paso a paso.  


## **1. Resolución del problema**

La situación indica:

- 8 obreros → 20 días para hacer una obra.
- Se pregunta: 2 obreros → ¿cuántos días?

Trabajan todos con la misma eficiencia.  
Número de obreros y tiempo son **inversamente proporcionales**, pues si hay menos obreros, tardan más tiempo.



Relación:  
$$
Obreros \times Días = constante
$$

$$
8 \times 20 = 160
$$



Si son 2 obreros:  
$$
2 \times Días = 160 \implies Días = \frac{160}{2} = 80
$$

**Respuesta correcta: (d) 80**

---



## **2. Estrategia para resolver en menos de 2 minutos**

- Identificar que es **inversamente proporcional**.
- Calcular la constante (obra total en obrero-días).
- Dividir esa constante por los nuevos obreros para hallar los días.
- Si fuese directamente proporcional, harías regla de tres directa; pero al ser inversa, usas producto constante.



También puedes aplicar regla de tres inversa:  
$$
8 \text{ obreros} \to 20 \text{ días}
$$ 
$$
2 \text{ obreros} \to x \text{ días}
$$  
Inversa: $ 8 \times 20 = 2 \times x \implies x = 80 $

# Tema 5: Problema 2  

<img src = 'pregunta_2.jpeg' width = 200>

Para resolver el problema de la imagen **pregunta_2.jpeg** sobre segmentos de recta, aplicaremos un enfoque algebraico sistemático. Aquí tienes la edición, el modelado y la resolución interactiva.

---



### 1. Enunciado del Problema (5.2)

En el segmento $AD$ ilustrado en la figura, se conocen las siguientes medidas y relaciones:
* La distancia entre los puntos $A$ y $C$ es $AC = 18$.
* La distancia entre los puntos $B$ y $D$ es $BD = 23$.
* El segmento $CD$ mide el doble que el segmento $AB$. Es decir, $CD = 2 \cdot AB$.

**El objetivo es hallar el valor total del segmento $AD$.**

---

### 2. Estrategia Relámpago (< 2 min)

Para resolver esto bajo presión, definamos variables para los segmentos elementales:
* Sea $x = AB$
* Entonces $CD = 2x$
* Sea $y = BC$

A partir de la imagen, planteamos un sistema de dos ecuaciones simples:
1.  $AC = x + y = 18$
2.  $BD = y + 2x = 23$

**Resolución mental rápida:**
Resta la primera ecuación de la segunda:
$$(y + 2x) - (x + y) = 23 - 18 \implies x = 5$$
Si $x = 5$, entonces $y = 18 - 5 = 13$.
El segmento total es $AD = x + y + 2x = 3x + y = 3(5) + 13 = 15 + 13 = 28$.

**Respuesta: (b) 28**

---

### 3. Modelado Matemático con SymPy

Utilizamos `SymPy` para validar la solución simbólicamente.


````python
import sympy as sp

# Definición de variables
x, y = sp.symbols('x y') # x = AB, y = BC

# Ecuaciones basadas en el enunciado
eq1 = sp.Eq(x + y, 18)      # AC = 18
eq2 = sp.Eq(y + 2*x, 23)    # BD = 23 (donde CD = 2x)

# Resolver el sistema
solucion = sp.solve((eq1, eq2), (x, y))

# Calcular AD = AB + BC + CD = x + y + 2x
ab = solucion[x]
bc = solucion[y]
cd = 2 * ab
ad_total = ab + bc + cd

print(f"AB: {ab}, BC: {bc}, CD: {cd}")
print(f"Valor de AD: {ad_total}")
````


### 4. Simulación Interactiva con ipywidgets

### Resumen de Fórmulas
Para cualquier punto $B$ y $C$ entre $A$ y $D$, se cumple:
$$AD = AB + BC + CD$$
Dado que $AC = AB + BC$ y $BD = BC + CD$, podemos expresar:
$$AD = AC + BD - BC$$
Sustituyendo los valores hallados:
$$AD = 18 + 23 - 13 = 28$$

Vamos a resolver el problema paso a paso.

---

$$

$$

$$



## **3. Calcular AD**

$ AB = 5 $, \( CD = 10 \), $BC = 18 - 5 = 13 $.

$$
AD = AB + BC + CD = 5 + 13 + 10 = 28
$$

**Respuesta correcta: (b) 28**

# Tema 5 Pregunta 3 del Tema Relaciones Directa e Inversamente Proporcionales 

<img src = 'pregunta_3.jpeg' width = 200>

Para resolver el problema **5.3** de la imagen `pregunta_3.jpeg` sobre el cálculo de porcentajes de forma eficiente, utilizaremos una combinación de razonamiento mental rápido y herramientas de computación científica para profundizar en el concepto.

---



### 1. Enunciado del Problema

> "5.3. El **12%** de **8000 pesos** es:"

* **(a)** 800
* **(b)** 960
* **(c)** 1200
* **(d)** 1460

---



### 2. Estrategia de Resolución Relámpago (< 10 segundos)

En un examen de admisión, no necesitas papel y lápiz para esto si aplicas la técnica de **Descomposición Decimal**:

1.  **Halla el 10%:** Simplemente quita un cero (divide por 10). El $10\%$ de $8000$ es **$800$**.
2.  **Halla el 1%:** Quita dos ceros (divide por 100). El $1\%$ de $8000$ es **$80$**.
3.  **Calcula el 2%:** Es el doble del $1\%$, es decir, **$160$**.
4.  **Suma (10% + 2%):** $800 + 160 = 960$.

**Respuesta Correcta: (b) 960**.

---



### 3. Modelado Matemático con SymPy

Para garantizar la precisión simbólica y manejar porcentajes más complejos, podemos definir una función de porcentaje en `SymPy`.


````python
import sympy as sp

monto, porcentaje = sp.symbols('monto porcentaje')
formula_porcentaje = (porcentaje / 100) * monto

# Sustitución de valores
resultado = formula_porcentaje.subs({porcentaje: 12, monto: 8000})
print(f"El resultado simbólico es: {resultado}")
````

### 4. Visualización y Experimentación (Pandas y ipywidgets)

Para fomentar el **aprendizaje por observación**, este componente interactivo permite visualizar qué parte del "todo" representa el porcentaje seleccionado.



````python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets

def calcular_y_graficar(pct, total):
    valor = (pct / 100) * total
    restante = total - valor
    
    # Crear DataFrame para Pandas
    df = pd.DataFrame({
        'Categoría': [f'{pct}% Calculado', 'Restante'],
        'Valores': [valor, restante]
    })
    
    # Graficar con Matplotlib
    plt.figure(figsize=(6, 6))
    colors = ['#ff9999','#66b3ff']
    plt.pie(df['Valores'], labels=df['Categoría'], autopct=lambda p: f'{p*total/100:.0f} pesos', 
            startangle=90, colors=colors, explode=(0.1, 0))
    plt.title(f"Visualización: {pct}% de {total}")
    plt.savefig(r"C:\Users\marco\Downloads\porcentaje_grafico.png")  # Guardar la figura como imagen
    plt.show()

# Widget interactivo
widgets.interact(calcular_y_graficar, 
                 pct=widgets.IntSlider(min=1, max=100, step=1, value=12),
                 total=widgets.fixed(8000))
````

### Resumen de la Operación
La expresión matemática formal para este cálculo es:

$$V = \frac{12}{100} \cdot 8000$$

Simplificando los ceros en la fracción y el entero:

$$V = 12 \cdot 80$$
$$V = 960$$

Esta técnica de **simplificación de ceros** es la estrategia más robusta para resolver problemas de este tipo en menos de 2 minutos sin errores de cálculo.

# Tema 5 Problema 4

<img src = 'pregunta_4.jpeg' width = 100>

Vamos paso a paso.

---

## **1. Interpretación del problema**

- Total inicial: $A + B = 2000$
- Del local $A$ se transporta el **20% de lo almacenado en A** al local $B$.
- Después del traslado, ambos locales tienen **la misma cantidad**.

Queremos $A$ inicial.

---



## **2. Plantear ecuaciones**

Inicial:  
$$
A + B = 2000
$$

Transporte: se mueve \( 0.2A \) sacos de \( A \) a \( B \).



Después del traslado:

$$
A' = A - 0.2A = 0.8A
$$
$$
B' = B + 0.2A
$$



Condición final:  
\[
A' = B'
\]
$$
0.8A = B + 0.2A
$$

---



## **3. Resolver**

De la ecuación de condición final:

$$
0.8A - 0.2A = B
$$
$$
0.6A = B
$$



Usando \( A + B = 2000 \):

$$
A + 0.6A = 2000
$$
$$
1.6A = 2000
$$
$$
A = \frac{2000}{1.6} = 1250
$$

Entonces:  
$$
A = 1250, \quad B = 750
$$

---



**Comprobación:**

- 20% de A = \( 0.2 \times 1250 = 250 \) sacos se pasan a B.
- Final: $ A = 1250 - 250 = 1000 $, $B = 750 + 250 = 1000$.

Sí, iguales.

**Respuesta correcta: (d) 1250**

---



## **4. Representación gráfica con matplotlib**

Podemos mostrar:

- Barras iniciales: A = 1250, B = 750.

## **1. Planteamiento del problema**

**Datos:**

- Distancia \( d \) (km) misma para A, B, C.
- Velocidades:  
  - B: \( v_B = 60 \) km/h  
  - C: \( v_C = 70 \) km/h  
  - A: \( v_A = ? \)

- Tiempos (en horas):  
  - \( t_B = t_A + \frac{4}{60} \) h (4 minutos más que A)  
  - \( t_C = t_A - \frac{2}{60} \) h (2 minutos menos que A)

Relación: tiempo = distancia / velocidad.  

$$
t_A = \frac{d}{v_A}, \quad t_B = \frac{d}{60}, \quad t_C = \frac{d}{70}
$$

---



## **2. Ecuaciones**

De \( t_B = t_A + \frac{4}{60} \):

\[
\frac{d}{60} = \frac{d}{v_A} + \frac{1}{15} \quad (1)
\]

De \( t_C = t_A - \frac{2}{60} \):

\[
\frac{d}{70} = \frac{d}{v_A} - \frac{1}{30} \quad (2)
\]

---



## **3. Resolver sistema**

De (2):  
\[
\frac{d}{v_A} = \frac{d}{70} + \frac{1}{30}
\]

Sustituir en (1):  
\[
\frac{d}{60} = \frac{d}{70} + \frac{1}{30} + \frac{1}{15}
\]
\[
\frac{d}{60} - \frac{d}{70} = \frac{1}{30} + \frac{1}{15}
\]
\[
d \left( \frac{7 - 6}{420} \right) = \frac{1 + 2}{30}
\]
\[
\frac{d}{420} = \frac{3}{30} = \frac{1}{10}
\]
\[
d = 42 \ \text{km}
\]

---



**5.5 Respuesta: (a) 42**

De (2):  
\[
\frac{42}{v_A} = \frac{42}{70} + \frac{1}{30}

\[
\frac{42}{v_A} = 0.6 + 0.0333\ldots = 0.6333\ldots
\]
\[
v_A = \frac{42}{0.6333\ldots} \approx 66.315\ \text{km/h}

```python
categorias = ["Aprobados","Reprobados"]
valores = [72,28]

plt.bar(categorias,valores)
plt.title("Resultados del examen")
plt.ylabel("Número de estudiantes")
plt.show()
```

---

## Pregunta

Según el gráfico, ¿qué porcentaje de estudiantes **aprobó**?

A) 65%
B) 70%
C) 72%
D) 80%

---

## Paso 1: calcular total de estudiantes

$$
72 + 28 = 100
$$

---

## Paso 2: calcular porcentaje

$$
\frac{72}{100}\times100
$$

---

$$
72%
$$

---

✅ **Respuesta correcta: C**

# Problema 2

# Crecimiento de población

El gráfico muestra la población de una ciudad.

```python
import matplotlib.pyplot as plt

años = [2020,2024]
poblacion = [10000,12000]

plt.plot(años,poblacion,marker="o")
plt.title("Crecimiento poblacional")
plt.ylabel("Habitantes")
plt.show()
```

---

## Pregunta

¿Cuál fue el **porcentaje de crecimiento** de la población?

A) 10%
B) 15%
C) 20%
D) 25%

---

## Paso 1: aumento

$$
12000 - 10000 = 2000
$$

---

## Paso 2: porcentaje de aumento

$$
\frac{2000}{10000}\times100
$$

---

$$
20%
$$

---

✅ **Respuesta correcta: C**
