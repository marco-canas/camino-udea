## **1. Resolución del problema**

La situación indica:

- 8 obreros → 20 días para hacer una obra.
- Se pregunta: 2 obreros → ¿cuántos días?

Trabajan todos con la misma eficiencia.  
Número de obreros y tiempo son **inversamente proporcionales**, pues si hay menos obreros, tardan más tiempo.

Relación:  
\[
Obreros \times Días = constante
\]

\[
8 \times 20 = 160
\]

Si son 2 obreros:  
\[
2 \times Días = 160 \implies Días = \frac{160}{2} = 80
\]

**Respuesta correcta: (d) 80**

---

## **2. Estrategia para resolver en menos de 2 minutos**

- Identificar que es **inversamente proporcional**.
- Calcular la constante (obra total en obrero-días).
- Dividir esa constante por los nuevos obreros para hallar los días.
- Si fuese directamente proporcional, harías regla de tres directa; pero al ser inversa, usas producto constante.

También puedes aplicar regla de tres inversa:  
\[
8 \text{ obreros} \to 20 \text{ días}
\]  
\[
2 \text{ obreros} \to x \text{ días}
\]  
Inversa: \( 8 \times 20 = 2 \times x \implies x = 80 \)

---

## **3. Representación con matplotlib y plotly**

### Función de proporcionalidad inversa:
\[
y = \frac{k}{x}
\]
donde \(x\) es número de obreros, \(y\) días, \(k = 160\).

```python


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Datos
k = 160
x = np.linspace(1, 16, 300)  # obreros
y = k / x

# --- Matplotlib ---
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Días = 160 / Obreros', color='b')
plt.scatter([8, 2], [20, 80], color='red', zorder=5)
plt.text(8, 20, '  (8,20)', fontsize=12, verticalalignment='bottom')
plt.text(2, 80, '  (2,80)', fontsize=12, verticalalignment='bottom')
plt.xlabel('Número de obreros')
plt.ylabel('Días necesarios')
plt.title('Proporcionalidad inversa: Obreros vs Días')
plt.grid(True)
plt.legend()
plt.show()
````

````python

# --- Plotly ---
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Días = 160 / Obreros'))
fig.add_trace(go.Scatter(x=[8, 2], y=[20, 80], mode='markers+text',
                         text=['(8,20)', '(2,80)'], textposition='top center',
                         marker=dict(color='red', size=10), name='Puntos del problema'))
fig.update_layout(title='Proporcionalidad inversa (Plotly)',
                  xaxis_title='Obreros',
                  yaxis_title='Días')
fig.show()
fig.write_image(r"C:\Users\marco\Downloads\proporcionalidad_inversa.png")
````

# Tema 5 Problema 3 

<img src = 'tema_5_problema_3.jpeg' width = 600>

Vamos paso a paso.

---

## **1. Interpretación del problema**

- Total inicial: $A + B = 2000$
- Del local $A$ se transporta el **20% de lo almacenado en A** al local $B$.
- Después del traslado, ambos locales tienen **la misma cantidad**.

Queremos $A$ inicial.

---



## **2. Plantear ecuaciones**

Inicial:  
\[
A + B = 2000
\]

Transporte: se mueve \( 0.2A \) sacos de \( A \) a \( B \).

Después del traslado:

\[
A' = A - 0.2A = 0.8A
\]
\[
B' = B + 0.2A
\]

Condición final:  
\[
A' = B'
\]
\[
0.8A = B + 0.2A
\]

---

## **3. Resolver**

De la ecuación de condición final:

\[
0.8A - 0.2A = B
\]
\[
0.6A = B
\]

Usando \( A + B = 2000 \):

$$
A + 0.6A = 2000
$$
$$
1.6A = 2000
$$
$$
A = \frac{2000}{1.6} = 1250
$$

Entonces:  
$$
A = 1250, \quad B = 750
$$

---



**Comprobación:**

- 20% de A = \( 0.2 \times 1250 = 250 \) sacos se pasan a B.
- Final: $ A = 1250 - 250 = 1000 $, $B = 750 + 250 = 1000$.

Sí, iguales.

**Respuesta correcta: (d) 1250**

---



## **4. Representación gráfica con matplotlib**

Podemos mostrar:

- Barras iniciales: A = 1250, B = 750.
- Flecha indicando traslado de 250 sacos.
- Barras finales: A = 1000, B = 1000.


````python
import matplotlib.pyplot as plt
import numpy as np

# Datos
A_inicial = 1250
B_inicial = 750
transporte = 0.2 * A_inicial
A_final = A_inicial - transporte
B_final = B_inicial + transporte

categorias = ['Local A', 'Local B']
inicial = [A_inicial, B_inicial]
final = [A_final, B_final]

x = np.arange(len(categorias))
width = 0.35

fig, ax = plt.subplots(figsize=(8,6))
rects1 = ax.bar(x - width/2, inicial, width, label='Inicial', color='skyblue')
rects2 = ax.bar(x + width/2, final, width, label='Final', color='lightgreen')

# Flecha de traslado
ax.annotate('', xy=(0.15, A_inicial - transporte/2), xytext=(0.15, A_inicial),
            arrowprops=dict(arrowstyle='->', color='red', lw=2))
ax.text(0.2, A_inicial - transporte/2, f'{transporte} sacos', color='red', ha='center')

ax.set_ylabel('Sacos de azúcar')
ax.set_title('Cantidad de sacos en A y B')
ax.set_xticks(x)
ax.set_xticklabels(categorias)
ax.legend()

# Valores en las barras
for i, (v1, v2) in enumerate(zip(inicial, final)):
    ax.text(i - width/2, v1 + 20, str(v1), ha='center')
    ax.text(i + width/2, v2 + 20, str(v2), ha='center')

plt.tight_layout()
plt.savefig(r'C:\Users\marco\Downloads\cantidad_sacos_A_y_B.png')
plt.show()
````

## **1. Planteamiento del problema**

**Datos:**

- Distancia \( d \) (km) misma para A, B, C.
- Velocidades:  
  - B: \( v_B = 60 \) km/h  
  - C: \( v_C = 70 \) km/h  
  - A: \( v_A = ? \)

- Tiempos (en horas):  
  - \( t_B = t_A + \frac{4}{60} \) h (4 minutos más que A)  
  - \( t_C = t_A - \frac{2}{60} \) h (2 minutos menos que A)

Relación: tiempo = distancia / velocidad.  

\[
t_A = \frac{d}{v_A}, \quad t_B = \frac{d}{60}, \quad t_C = \frac{d}{70}
\]

---

## **2. Ecuaciones**

De \( t_B = t_A + \frac{4}{60} \):

\[
\frac{d}{60} = \frac{d}{v_A} + \frac{1}{15} \quad (1)
\]

De \( t_C = t_A - \frac{2}{60} \):

\[
\frac{d}{70} = \frac{d}{v_A} - \frac{1}{30} \quad (2)
\]

---

## **3. Resolver sistema**

De (2):  
\[
\frac{d}{v_A} = \frac{d}{70} + \frac{1}{30}
\]

Sustituir en (1):  
\[
\frac{d}{60} = \frac{d}{70} + \frac{1}{30} + \frac{1}{15}
\]
\[
\frac{d}{60} - \frac{d}{70} = \frac{1}{30} + \frac{1}{15}
\]
\[
d \left( \frac{7 - 6}{420} \right) = \frac{1 + 2}{30}
\]
\[
\frac{d}{420} = \frac{3}{30} = \frac{1}{10}
\]
\[
d = 42 \ \text{km}
\]

---

**5.5 Respuesta: (a) 42**

---

## **4. Velocidad de A**

De (2):  
\[
\frac{42}{v_A} = \frac{42}{70} + \frac{1}{30}
\]
\[
\frac{42}{v_A} = 0.6 + 0.0333\ldots = 0.6333\ldots
\]
\[
v_A = \frac{42}{0.6333\ldots} \approx 66.315\ \text{km/h}
\]

---

**5.6 Respuesta: (b) 66.3**

---

## **5. Gráfico con matplotlib**

- \( t_C = 42 / 70 = 0.6\ \text{h} = 36\ \text{min} \)

```python

**Problema 3:**

# Concepto fundamental

El porcentaje representa una proporción de cada cien.

$$
\text{porcentaje}=\frac{\text{parte}}{\text{total}}\times100
$$

---

# Estrategia general para el examen

1. Identificar **total**
2. Identificar **parte**
3. Aplicar proporción

$$
\frac{\text{parte}}{\text{total}}
