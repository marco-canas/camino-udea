# Concepto fundamental

El porcentaje representa una proporción de cada cien.

$$
\text{porcentaje}=\frac{\text{parte}}{\text{total}}\times100
$$

---

# Estrategia general para el examen

1. Identificar **total**
2. Identificar **parte**
3. Aplicar proporción

$$
\frac{\text{parte}}{\text{total}}
$$

# Problema 1

Un producto cuesta **$200000**.

Tiene un descuento del **15%**.

¿Cuál es el precio final?

A) 160000
B) 170000
C) 180000
D) 185000

---

## Estrategia

Calcular el valor del descuento.

$$
0.15 \times 200000
$$

---

## Cálculo

$$
30000
$$

Precio final:

$$
200000-30000=170000
$$

Respuesta correcta: **B**

---

## Python

```python
precio=200000
descuento=0.15

precio_final=precio*(1-descuento)

precio_final
```

# Problema 5

Un salario aumenta **10%** y luego **10%** nuevamente.

¿Cuál es el aumento total aproximado?

A) 20%
B) 21%
C) 22%
D) 25%

---

## Estrategia

Si el salario inicial es $100$:

Primer aumento:

$$
100(1.1)=110
$$

Segundo aumento:

$$
110(1.1)=121
$$

---

Aumento total:

$$
121-100=21%
$$

Respuesta **B**

---

## Python

```python
100*1.1*1.1
```

# Problema 1

# Resultados de un examen

El gráfico muestra los resultados de un examen en un grupo de estudiantes.

### Código para generar el gráfico

```python
import matplotlib.pyplot as plt

categorias = ["Aprobados","Reprobados"]
valores = [72,28]

plt.bar(categorias,valores)
plt.title("Resultados del examen")
plt.ylabel("Número de estudiantes")
plt.show()
```

---

## Pregunta

Según el gráfico, ¿qué porcentaje de estudiantes **aprobó**?

A) 65%
B) 70%
C) 72%
D) 80%

---

## Paso 1: calcular total de estudiantes

$$
72 + 28 = 100
$$

---

## Paso 2: calcular porcentaje

$$
\frac{72}{100}\times100
$$

---

$$
72%
$$

---

✅ **Respuesta correcta: C**
