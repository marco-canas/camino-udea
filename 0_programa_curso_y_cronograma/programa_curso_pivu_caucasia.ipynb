# Programa del curso de Camino a la UdeA o PIVU Puerto Libertador (Bijao)  

Esta es la actualización del programa para el municipio de **Puerto Libertador, Córdoba**. Para cumplir con la intensidad horaria de **48 horas** totales, el curso se desarrollará durante **8 sábados consecutivos**, dedicando **2 sábados (12 horas)** a cada uno de los 4 bloques temáticos principales.

Las sesiones serán de **6 horas** (por ejemplo, de 8:00 AM a 3:00 PM con una hora de receso), permitiendo cubrir el diagnóstico inicial y el simulacro final en los extremos del cronograma.




## Cronograma: Camino a la UdeA - Puerto Libertador
**Lugar:** Puerto Libertador, Córdoba.
**Horario:** Sábados (6 horas por sesión).
**Total:** 8 sesiones (48 horas).

### **Módulo I: Lenguaje Proposicional**
* **Sábado 11 de abril:**
    * **Bloque 1 (1.5 h):** Prueba de diagnóstico inicial.
    * **Bloque 2 (4.5 h):** Introducción al análisis del lenguaje y conectivos lógicos.
* **Sábado 18 de abril:**
    * **Bloque 1 (4 h):** Control de variables y posibilidades lógicas.
    * **Bloque 2 (2 h):** **Evaluación I: Lenguaje Proposicional**.

### **Módulo II: Esquemas de Proporcionalidad**
* **Sábado 25 de abril:**
    * **Contenido (6 h):** Relaciones directas e inversas, y porcentajes.
* **Sábado 2 de mayo:**
    * **Bloque 1 (4 h):** Compensaciones aditivas, multiplicativas y probabilidad.
    * **Bloque 2 (2 h):** **Evaluación II: Proporcionalidad**.

### **Módulo III: Razonamiento Abstracto**
* **Sábado 9 de mayo:**
    * **Contenido (6 h):** Operaciones no convencionales y razonamiento aritmético.
* **Sábado 16 de mayo:**
    * **Bloque 1 (4 h):** Series numéricas, alfanuméricas y patrones lógicos.
    * **Bloque 2 (2 h):** **Evaluación III: Razonamiento Abstracto**.

### **Módulo IV: Razonamiento Geométrico**
* **Sábado 23 de mayo:**
    * **Contenido (6 h):** Áreas, perímetros, volúmenes y relaciones espaciales.
* **Sábado 30 de mayo:**
    * **Bloque 1 (2.5 h):** Series gráficas y análisis de datos en gráficos.
    * **Bloque 2 (2 h):** **Evaluación IV: Geometría**.
    * **Bloque 3 (1.5 h):** Prueba final y cierre del curso.





## Estructura Pedagógica por Sábado (6 Horas)
Para mantener el enfoque intensivo, cada jornada se dividirá así:
* **Fase Conceptual (2 h):** Explicación de estructuras lógicas y patrones.
* **Fase Práctica (2 h):** Resolución de ejercicios tipo examen con estrategias de eliminación.
* **Fase de Entrenamiento (2 h):** Taller cronometrado o evaluación programada (según el sábado).
