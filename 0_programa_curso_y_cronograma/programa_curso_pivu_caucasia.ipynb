# Estrategias didácticas del curso

### 1. Aprendizaje basado en patrones

El examen evalúa **reconocimiento de estructuras lógicas** más que cálculo.

### 2. Resolución rápida

Se entrenan **estrategias de eliminación de opciones**.

### 3. Simulacros cronometrados

Para desarrollar:

* velocidad
* concentración
* control del tiempo



### 4. Análisis de errores

Después de cada simulacro se estudia:

* tipo de error
* estrategia correcta
