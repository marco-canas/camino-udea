# Preguntas basadas en el gráfico de torta de inversiones.   

Estas preguntas son ideales para aplicar **razonamiento proporcional rápido**.



## 1. Transcripción y Edición de los Problemas

**Contexto General:**
El gráfico de sectores (torta) muestra la distribución de una inversión total de $108.000$ US en 4 departamentos (A, B, C, D). Se sabe que:
* Departamento **D**: $34\%$
* Departamento **B**: $9\%$
* La suma de las inversiones en **B** y **C** es de $28.080$ US.

### Pregunta 54
**La inversión realizada en el departamento A, en US, fue de:**
(A) $50.000$
(B) $43.200$
(C) $41.300$
(D) $55.000$

### Pregunta 55
**El porcentaje del total que representa la inversión en C es:**
(A) $17\%$
(B) $19\%$
(C) $21\%$
(D) $23\%$

---



## 2. Resolución Didáctica (Estrategias de < 2 min)

### Para la Pregunta 55 (Resolver esta primero ayuda a la 54)
**Estrategia: Conversión de monto a porcentaje.**
Sabemos que el $100\%$ es $108.000$ US. Podemos hallar qué porcentaje representan los $28.080$ US de la suma de B y C:
$$\%_{(B+C)} = \frac{28.080}{108.000} \times 100$$
Simplificando (tachando ceros): $\frac{2808}{108}$. 
Si dividimos $2808 \div 108$, obtenemos exactamente **$26\%$**.

Como sabemos que **B** es el $9\%$:
$$\%_C = 26\% - 9\% = 17\%$$
**Respuesta 55: (A) 17%**

---



### Para la Pregunta 54
**Estrategia: Complemento al 100%.**
Ahora que conocemos los porcentajes de B, C y D, el departamento A debe ser el resto para completar el círculo ($100\%$):
$$\%_A = 100\% - (D\% + B\% + C\%)$$
$$\%_A = 100\% - (34\% + 9\% + 17\%) = 100\% - 60\% = 40\%$$

Calculamos el $40\%$ del total ($108.000$):
$$Inversión_A = 0.40 \times 108.000 = 43.200$$
**Respuesta 54: (B) 43.200**

---



## 3. Representación de Datos y Visualización

Utilizaremos **Pandas** y **Plotly** para recrear el gráfico y verificar los valores.


### Tabla de Inversiones (Pandas)


```python
import pandas as pd
import plotly.express as px

# Datos calculados
total_inv = 108000
data = {
    'Departamento': ['A', 'B', 'C', 'D'],
    'Porcentaje (%)': [40, 9, 17, 34]
}

df = pd.DataFrame(data)
df['Monto (US)'] = (df['Porcentaje (%)'] / 100) * total_inv

print(df.to_string(index=False))

# Gráfico Interactivo con Plotly
fig = px.pie(df, values='Monto (US)', names='Departamento', 
             title='Distribución de Inversión Total (108,000 US)',
             hover_data=['Porcentaje (%)'],
             color_discrete_sequence=px.colors.qualitative.Set3)
fig.update_traces(textinfo='percent+label')
fig.show()
fig.write_image("C:\\Users\\marco\\Downloads\\distribucion_inversion_departamentos.png")
```

### Visualización de Montos con Matplotlib


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
bars = plt.bar(df['Departamento'], df['Monto (US)'], color='skyblue')

# Añadir etiquetas de monto sobre las barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500, f'${int(yval):,}', ha='center', va='bottom')

plt.title('Monto de Inversión por Departamento')
plt.ylabel('Dólares (US)')
plt.ylim(0, 50000)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("C:\\Users\\marco\\Downloads\\inversion_departamentos.png", bbox_inches='tight')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Preparación de los datos (basado en los cálculos anteriores)
data = {
    'Departamento': ['A', 'B', 'C', 'D'],
    'Porcentaje (%)': [40, 9, 17, 34]
}
df = pd.DataFrame(data)

# 2. Creación del gráfico de barras por porcentaje
plt.figure(figsize=(10, 6))
colores = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2'] # Colores distintivos

# Creamos las barras usando la columna de Porcentaje
barras = plt.bar(df['Departamento'], df['Porcentaje (%)'], color=colores, edgecolor='black', alpha=0.8)

# 3. Personalización y Estética
plt.title('Distribución Porcentual de la Inversión por Departamento', fontsize=14, pad=20)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Porcentaje del Total (%)', fontsize=12)
plt.ylim(0, 50)  # Damos un poco de espacio sobre la barra más alta (40%)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Añadir etiquetas de datos exactos sobre cada barra
for barra in barras:
    alto = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, alto + 1, 
             f'{alto}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.savefig("C:\\Users\\marco\\Downloads\\porcentaje_inversion_departamentos.png", bbox_inches='tight')
plt.show()
```
