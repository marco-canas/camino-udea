Este es un reto clásico de **lógica de declaraciones (verdades y mentiras)**. 

Para resolverlo en menos de dos minutos, utilizaremos la técnica de **Identificación de Contradicciones**.

---



## 1. Transcripción y Edición del Problema

**Enunciado:**
Ana, Lina y María sostienen la siguiente conversación:
* **Ana:** "Yo soy menor de edad".
* **Lina:** "Ana miente".
* **María:** "Lina es mayor de edad".

**Condiciones del problema:**
1.  **Solo una** de ellas miente (las otras dos dicen la verdad).
2.  **Solo una** de ellas es mayor de edad (las otras dos son menores).

**Pregunta:**
¿Quién es la que miente y quién es la mayor de edad, respectivamente?
(A) Lina - Lina
(B) Lina - María
(C) María - Lina
(D) Ana - Lina




## 2. Resolución Didáctica (Estrategia: El Duelo de Contradicciones)

**Paso 1: Identificar la contradicción.**
Observa las declaraciones de **Ana** y **Lina**:
* Ana dice: "Soy menor".
* Lina dice: "Ana miente" (lo que equivale a decir: "Ana es mayor") o lo que es lo mismo que decir que "Ana no es menor".

Como una afirma lo contrario de la otra, **una de las dos debe estar mintiendo y la otra diciendo la verdad**.



**Paso 2: Determinar la veracidad de María.**
El problema dice que **solo hay una mentirosa**. Si la mentirosa está entre Ana y Lina, entonces **María debe decir la verdad forzosamente**.
* Declaración de María (Verdad): "Lina es mayor de edad".



**Paso 3: Aplicar la segunda condición (Solo hay una mayor de edad).**
Si María dice la verdad, **Lina es la mayor de edad**. Como solo puede haber una mayor de edad, Ana y María deben ser menores.



**Paso 4: Verificar quién miente.**
* Ana dijo: "Soy menor". Como ya sabemos que es menor, Ana **dice la verdad**.
* Lina dijo: "Ana miente". Como Ana dijo la verdad, entonces **Lina es la que miente**.

**Resultado:**
* La que miente: **Lina**.
* La que es mayor de edad: **Lina**.

**Respuesta Correcta: (A) Lina - Lina.**

## 3. Representación de Datos y Scripts en Python

Para visualizar este razonamiento de forma tabular y gráfica, utilizaremos las librerías solicitadas.



### Análisis Tabular de Escenarios (Pandas)
Este script evalúa la coherencia de las declaraciones basándose en quién asumimos que miente.


```python
import pandas as pd

def analizar_logica():
    # Escenarios: ¿Quién miente?
    escenarios = ['Ana miente', 'Lina miente', 'María miente']
    resultados = []

    for quien_miente in escenarios:
        # Definimos veracidad según el escenario
        v_ana = quien_miente != 'Ana miente'
        v_lina = quien_miente != 'Lina miente'
        v_maria = quien_miente != 'María miente'
        
        # Deducciones según las frases:
        # María: Lina es mayor (si María es veraz)
        es_mayor = "Lina" if v_maria else "No Lina"
        
        # Ana: Yo soy menor (Si Ana es veraz, Ana es menor. Si Ana miente, Ana es mayor)
        ana_condicion = "Menor" if v_ana else "Mayor"
        
        # Lina: Ana miente (Lina debe ser veraz si su frase coincide con el estado de Ana)
        lina_es_veraz_real = (v_ana == False)
        coherencia_lina = (v_lina == lina_es_veraz_real)

        resultados.append({
            'Supuesto: Miente': quien_miente.split()[0],
            'Deducción Mayoría': es_mayor,
            'Estado Ana': ana_condicion,
            '¿Coherente?': "Sí" if coherencia_lina and (v_maria) else "No"
        })

    return pd.DataFrame(resultados)

df = analizar_logica()
print(df.to_string(index=False))
```
