## 📝 Enunciado Editado y Completado
**Problema:** Martha y Leonor juegan una serie de 18 partidas de cartas. Acuerdan que, en cada partida, la perdedora debe pagar a la ganadora **$5 USD**. Si al finalizar todas las partidas Martha tiene una ganancia neta de **$40 USD**, ¿cuántas partidas ganó Leonor?

---

## 🚀 Estrategias para resolverlo en menos de 2 minutos

Para ganar tiempo en un examen, no uses sistemas de ecuaciones complejos. Usa uno de estos dos métodos:

### Método 1: El "Tanteo Inteligente" (Súper rápido)
1. **Punto de equilibrio:** Si hubieran empatado (9 partidas cada una), Martha tendría **$0**.
2. **Ajuste:** Como Martha ganó **$40**, ella debe tener más victorias que Leonor.
3. **Prueba una opción:** Si Leonor ganó **5** partidas (Opción C), entonces Martha ganó **13** ($18 - 5 = 13$).
   * Diferencia de partidas: $13 - 5 = 8$ partidas a favor de Martha.
   * Ganancia: $8 \text{ partidas} \times \$5 = \$40$.
   * **¡Listo! Coincide perfectamente.**

### Método 2: La Lógica de la Diferencia (Fórmula mental)
1. **Ganancia total / Valor por partida:** $\$40 / \$5 = 8$. Esto significa que Martha ganó **8 partidas más** que Leonor.
2. **Reparto del total:**
   * Tienes 18 partidas en total.
   * Resta la diferencia: $18 - 8 = 10$.
   * Divide ese resto entre las dos personas: $10 / 2 = 5$.
   * **Leonor ganó 5 partidas.** (Y Martha ganó $5 + 8 = 13$).

---

## ✍️ Resolución Didáctica (Paso a paso)

Si prefieres verlo de forma matemática tradicional, definimos:
* $M$: Partidas ganadas por Martha.
* $L$: Partidas ganadas por Leonor.

Sabemos que:
1.  $M + L = 18$ (Total de partidas)
2.  $5M - 5L = 40$ (Dinero ganado menos dinero perdido)

**Simplificamos la segunda ecuación dividiendo todo entre 5:**
$M - L = 8$

**Ahora sumamos las dos ecuaciones:**
$$(M + L) + (M - L) = 18 + 8$$
$$2M = 26$$
$$M = 13$$

**Calculamos las de Leonor:**
$$L = 18 - 13 = 5$$

---



## Resultado Final
El número de partidas que ganó Leonor fue:
**C. 5**

> **Tip Pro:** En este tipo de problemas de "gana/pierde", la diferencia de dinero siempre equivale a la **diferencia de partidas ganadas multiplicada por la apuesta**. Aprender esto te permite resolverlo mentalmente en menos de 30 segundos.
