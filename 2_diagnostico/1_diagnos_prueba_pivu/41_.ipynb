````python
import pandas as pd
import random

def estudiante_que_va_a_participar(ruta_archivo):
    """
    Lee un archivo Excel desde la hoja 'Inscritos' a partir de la fila 3
    y selecciona un estudiante al azar.
    
    Parámetros:
    ruta_archivo (str): Ruta del archivo .xlsx
    
    Retorna:
    str: Nombre del estudiante seleccionado al azar
    """
    try:
        # Leer el archivo Excel, hoja 'Inscritos', saltando las primeras 2 filas
        df = pd.read_excel(ruta_archivo, sheet_name="Inscritos", skiprows=2)
        
        # Verificar que la columna 'NOMBRES Y APELLIDOS' existe
        if 'NOMBRES Y APELLIDOS' not in df.columns:
            print(f"❌ Error: No se encontró la columna 'NOMBRES Y APELLIDOS'")
            print(f"Columnas disponibles: {df.columns.tolist()}")
            return None
        
        # Eliminar filas con valores nulos en la columna de nombres
        df_limpio = df[df['NOMBRES Y APELLIDOS'].notna()]
        
        # Verificar que hay al menos un estudiante
        if len(df_limpio) == 0:
            print(f"❌ Error: No hay estudiantes válidos en la lista")
            return None
        
        # Seleccionar un estudiante al azar
        estudiante_aleatorio = random.choice(df_limpio['NOMBRES Y APELLIDOS'].tolist())
        
        print(f"🎲 Estudiante seleccionado al azar:")
        print(f"✅ {estudiante_aleatorio}")
        
        return estudiante_aleatorio
        
    except FileNotFoundError:
        print(f"❌ Error: No se encontró el archivo en la ruta: {ruta_archivo}")
        return None
    except ValueError as e:
        print(f"❌ Error: No se encontró la hoja 'Inscritos'. {e}")
        return None
    except Exception as e:
        print(f"❌ Error inesperado: {e}")
        return None

# Uso del script
ruta = r"C:\Users\marco\Documentos\extension\camino-udea\9_asistencia_evaluacion\pivu_caucasia_notas_quices_asistencia.xlsx"
estudiante = estudiante_que_va_a_participar(ruta)

# Si solo quieres el nombre sin mensajes adicionales, usa esta versión más simple:

def estudiante_que_va_a_participar_simple(ruta_archivo):
    """Versión simple que solo retorna el nombre del estudiante"""
    try:
        df = pd.read_excel(ruta_archivo, sheet_name="Inscritos", skiprows=2)
        df_limpio = df[df['NOMBRES Y APELLIDOS'].notna()]
        return random.choice(df_limpio['NOMBRES Y APELLIDOS'].tolist())
    except:
        return None
````

# Pregunta número 41 de la prueba diagnóstica del PIVU

Anita recibe de su padre una bolsa con 5 bolitas de chicle de las cuales 3 son rojas y dos azules.   

Si Anita saca de la bolsa 3 bolas de chicle y se las come, entonces **se puede afirmar con certeza que**:   

A.  Anita se comió tres bolas de chicle del mismo color.     
B.  Anita se comió dos bolas de chicle azul y una roja.   
C. 	Anita se comió por lo menos una bola de chicle roja  
D.  Anita se comió por lo menos una bola de chicle azul  

## [Video de apoyo](https://www.youtube.com/watch?v=466RFF6jj4U)

Este es un problema clásico de **razonamiento lógico y el Principio del Palomar** (o de las casillas). En este tipo de ejercicios, la clave es identificar qué suceso es **obligatorio** que ocurra, independientemente de la suerte o el orden.

---

## 1. Transcripción y Edición del Problema

**Enunciado:**
Anita recibe de su padre una bolsa con $5$ bolitas de chicle, de las cuales $3$ son rojas y $2$ son azules. Si Anita saca de la bolsa $3$ bolas de chicle al azar y se las come, entonces **se puede afirmar con certeza que**:

(A) Anita se comió tres bolas de chicle del mismo color.
(B) Anita se comió dos bolas de chicle azul y una roja.
(C) Anita se comió por lo menos una bola de chicle roja.
(D) Anita se comió por lo menos una bola de chicle azul.

---

## 2. Resolución Didáctica (Estrategia de < 2 minutos)

Para resolver este problema con rapidez, debemos buscar el **"peor escenario posible"** para cada afirmación. Algo se afirma con "certeza" solo si no hay ninguna forma de que sea falso.

**Análisis de las opciones:**
1.  **¿Tres del mismo color? (A):** Como solo hay $2$ azules, la única forma de sacar tres del mismo color sería que las tres fueran rojas. ¿Es posible no sacar tres rojas? Sí, podría sacar $2$ rojas y $1$ azul. Por tanto, no hay certeza.
2.  **¿Dos azules y una roja? (B):** Esto es solo una posibilidad. También podría sacar $2$ rojas y $1$ azul, o $3$ rojas. No hay certeza.
3.  **¿Por lo menos una azul? (D):** ¿Existe un caso donde no saque ninguna azul? Sí. Si saca las $3$ rojas disponibles, se habrá comido $0$ azules. Por tanto, no hay certeza.
4.  **¿Por lo menos una roja? (C):** Intentemos el caso donde Anita "evite" las rojas. Ella solo puede sacar $2$ azules (porque no hay más). Como está obligada a sacar $3$ bolas, la tercera bola **tiene que ser roja** obligatoriamente.

**Conclusión:** No importa qué combinación elija, como máximo puede evitar las rojas en dos extracciones. La tercera siempre será roja.

**Respuesta Correcta:** (C) Anita se comió por lo menos una bola de chicle roja.

---

## 3. Representación y Cálculos con Python

A continuación, presento los scripts para modelar el espacio muestral y visualizar las probabilidades.

### Cálculo de Espacio Muestral (SymPy)
Usamos **SymPy** para listar todas las combinaciones posibles de 3 chicles.

```python


````python
from sympy.utilities.iterables import combinations

# Definimos la bolsa: R=Roja, A=Azul
bolsa = ['R1', 'R2', 'R3', 'A1', 'A2']

# Obtenemos todas las formas de sacar 3 chicles
comb = list(combinations(bolsa, 3))

print(f"Total de combinaciones posibles: {len(comb)}")
print("Combinaciones (R=Roja, A=Azul):")
for c in comb:
    print(c)
````

### Análisis Tabular (Pandas)
Utilizamos **Pandas** para verificar si en cada caso existe al menos una roja.




````python
import pandas as pd

# Analizamos el contenido de cada combinación
df_data = []
for c in comb:
    rojas = sum(1 for bola in c if 'R' in bola)
    azules = sum(1 for bola in c if 'A' in bola)
    df_data.append({
        'Combinación': str(c),
        'Rojas': rojas,
        'Azules': azules,
        '¿Al menos una Roja?': rojas >= 1,
        '¿Al menos una Azul?': azules >= 1
    })

df = pd.DataFrame(df_data)
print(df.to_string(index=False))
````

### Visualización de Probabilidades (Plotly & Matplotlib)


````python
import plotly.graph_objects as go

# Calculamos frecuencias de éxito para las afirmaciones C y D
p_roja = df['¿Al menos una Roja?'].mean() * 100
p_azul = df['¿Al menos una Azul?'].mean() * 100

fig = go.Figure(data=[
    go.Bar(name='Certeza', x=['Al menos una Roja', 'Al menos una Azul'], 
           y=[p_roja, p_azul], text=[f'{p_roja}%', f'{p_azul}%'],
           marker_color=['green', 'orange'])
])

fig.update_layout(title='Análisis de Certeza (Probabilidad del 100%)',
                  yaxis_title='Porcentaje de casos (%)',
                  template='plotly_white')
fig.show()
````

### Simulación de Monte Carlo (SciPy/NumPy)
Para confirmar de manera didáctica mediante experimentación masiva.


````python
import numpy as np

def simulacion_anita(n_intentos=10000):
    bolsa = np.array(['R', 'R', 'R', 'A', 'A'])
    exitos_roja = 0
    
    for _ in range(n_intentos):
        seleccion = np.random.choice(bolsa, 3, replace=False)
        if 'R' in seleccion:
            exitos_roja += 1
            
    return exitos_roja / n_intentos

prob_final = simulacion_anita()
print(f"En 10,000 simulaciones, la probabilidad de sacar al menos una roja es: {prob_final * 100}%")
````

**Resumen Didáctico Final:**
El razonamiento lógico nos dice que si tenemos un total de $N$ objetos de un tipo y queremos sacarlos todos pero el número de extracciones supera la cantidad de objetos del "otro" tipo, es imposible no sacar al menos uno del primero.
$$n_{ext} > n_{azules} \implies 3 > 2 \therefore \text{Certeza de Roja.}$$

## Solución: 

## Empecemos descartando la opción A. 

Veamos porqué no podemos estar seguros de la opción A, es decir, mostremos porqué no podemos afirmar con certeza que Anita se ha comido tres bolas de chiche del mismo color. 
De esta afirmación no podemos estar seguros pues al haber dos tipos de colores de bolitas de chicle, entonces al sacar tres podemos sacar
* todas del mismo color rojo, 
* o dos rojas, y una azul
* o una roja y dos azules
por lo tanto, no podemos estar seguros de que al sacar tres bolitas, tres sean del mismo color, pues sacar tres bolar rojas no es el único evento posible. 

## Sigamos ahora con el proceso de descarte de la opción B.  

Veamos porqué no podemos estar seguros de la opción B.  Es decir, veamos porqué no podemos estar seguros de que Anita se comió dos bolas de chicle azul y una roja. 

Porque también es posible que Anita se haya comido 3 bolitas de color rojo. 



### Veamos que la opción C es la correcta, 

es decir, veamos que podemos estar seguros de que Anita se comió por lo menos una bola de chicle roja.

De esto podemos estar seguros ya que al no haber sino solo dos bolas azules y comerse tre de las 5, obligatoriamente al menos una de las bolas debe ser roja. 

Utilizando la definición de probabilidad de un evento, veamos que el haberse comido por lo menos una bola de chicle roja es es evento de probabilidad del $100\%$

Empecemos por determinar el espacio muestral $S$ de este experimento aleatorio.

````python
#Simulación del experimento de sacar tres bolitas y comerselas
import numpy as np 
from sympy import *
a1, a2, r1, r2, r3 = symbols('a1, a2, r1, r2, r3')
bolsa = [a1, a2, r1, r2, r3]
selecciones = FiniteSet() # instanciaando un conjunto vacío para almacenar las selecciones

for i in range(3):
    a = np.random.randint(0, len(bolsa) - 1)
    selecciones += FiniteSet(bolsa[a]) # Agrega la bolita seleccionada al 
    #conjunto de selecciones
    bolsa.remove(bolsa[a])
   
    
    
print('Uno de los resultados de este experimento aleatorio es: ', selecciones) 
````
