**7.** La figura muestra una balanza de brazos iguales, la cual se encuentra en equilibrio cuando es cargada con pesas de magnitudes $A, B, C, X$ e $Y$; en las cantidades indicadas.


```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_balance(ax, left_blocks, right_blocks, title):
    # Dibuja la base de la balanza
    base = patches.Polygon([[-2, -1], [2, -1], [0, 0]], closed=True, facecolor='black')
    ax.add_patch(base)
    # Dibuja el brazo
    ax.plot([-3, 3], [0, 0], color='black', linewidth=3)
    
    # Función para dibujar una torre de bloques
    def draw_tower(x_pos, blocks, color):
        y_pos = 0
        block_height = 0.8
        block_width = 1.2
        for block_label in reversed(blocks): # De abajo hacia arriba
            rect = patches.Rectangle((x_pos - block_width/2, y_pos), block_width, block_height, 
                                     linewidth=1, edgecolor='black', facecolor=color)
            ax.add_patch(rect)
            ax.text(x_pos, y_pos + block_height/2, block_label, 
                    horizontalalignment='center', verticalalignment='center', fontsize=12, fontweight='bold')
            y_pos += block_height

    # Dibuja los bloques
    draw_tower(-2, left_blocks, 'lightgray')
    draw_tower(2, right_blocks, 'darkgray')

    # Configuración de los ejes
    ax.set_xlim(-4, 4)
    ax.set_ylim(-1.5, 4)
    ax.set_aspect('equal')
    ax.axis('off')
    ax.set_title(title, fontsize=14, pad=20)

# Crear la figura
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Definir los bloques para cada balanza (de abajo hacia arriba)
balance1_left = ['A', 'A', 'B', 'C']
balance1_right = ['X', 'X', 'X', 'Y']
balance2_left = ['A', 'B', 'X', 'Y']
balance2_right = ['A', 'C', 'X', 'X']

# Dibujar las balanzas
draw_balance(ax1, balance1_left, balance1_right, "Balanza 1: A + B = X + Y")
draw_balance(ax2, balance2_left, balance2_right, "Balanza 2: A + B = 2A")

plt.tight_layout()
plt.savefig(r"C:\Users\marco\Downloads\balanzas.png")  # Guardar la imagen
plt.show()

```
